To uncover any patterns in the purity of water sources over time, we want to determine during which months the testing procedures identified the highest and lowest number of water sources as 'clean'. 
Construct a SQL query to count the ‘clean’ water sources for each month - which months had the highest and lowest number of ‘clean’ water sources, respectively?

```python
%%sql
SELECT
    CAST(strftime('%m', date) AS INTEGER) AS month_num,
    CASE CAST(strftime('%m', date) AS INTEGER)
        WHEN 1  THEN 'January'   WHEN 2  THEN 'February'
        WHEN 3  THEN 'March'     WHEN 4  THEN 'April'
        WHEN 5  THEN 'May'       WHEN 6  THEN 'June'
        WHEN 7  THEN 'July'      WHEN 8  THEN 'August'
        WHEN 9  THEN 'September' WHEN 10 THEN 'October'
        WHEN 11 THEN 'November'  WHEN 12 THEN 'December'
    END AS month_name,
    COUNT(*) AS number_of_clean_sources
FROM well_pollution
WHERE results = 'Clean'
GROUP BY month_num, month_name
ORDER BY month_num;
```

What are the names of the two worst-performing employees who visited the fewest sites, and how many sites did the worst-performing employee visit? Modify your queries from the “Honouring the workers” section.

```python
%%sql
SELECT 
    e.employee_name,
    COUNT(v.visit_count) AS total_visits
FROM 
    visits v
JOIN 
    employee e ON v.assigned_employee_id = e.assigned_employee_id
GROUP BY 
    e.employee_name
ORDER BY 
    total_visits asc
LIMIT 5;
```

You're tasked with analysing the visits table to identify the queue status of visits. Write a query to select the location_id, a calculated total of visit_count, and create a new column named queue_status that classifies the time_in_queue of each visit as either 'Acceptable' (30 minutes or less) or 'Too Long' (more than 30 minutes).  

The query should count the number of visits and group them by their queue_status ('Acceptable' or 'Too Long') separately for each location. For example, if a location had a total of 5 visits, and, out of these, 3 visits had queue times of 30 minutes or less, while the remaining 2 visits had queue times exceeding 30 minutes,  the results should show:
<location_id>, 3, Acceptable
<location_id>, 2, Too Long

```python
%%sql
SELECT 
    location_id,
    COUNT(visit_count) AS total_visits,
    CASE 
        WHEN time_in_queue <= 30 THEN 'Acceptable'
        ELSE 'Too Long'
    END AS queue_status
FROM 
    visits
GROUP BY 
    location_id,
    queue_status
    LIMIT 10;
```

A citizen suspects a bias in the frequency of employee visits to water sources, potentially favouring certain days of the week over others. For our data to reflect an unbiased sampling, there should be a uniform distribution of visits across all days. Which SQL query would effectively test this hypothesis using the data from the visits table?

```python
%%sql
SELECT 
    DAYNAME(time_of_record) AS day_of_week,
    COUNT(*) AS total_visits
FROM 
    visits
GROUP BY 
    DAYNAME(time_of_record)
ORDER BY total_visits DESC
;
```

Which SQL query will produce the date format "DD Month YYYY" from the time_of_record column in the visits table, as a single column? Note: Monthname() acts in a similar way to DAYNAME().

```python
%%sql
SELECT CONCAT(day(time_of_record),
               " ",
                 monthname(time_of_record),
                   " ", year(time_of_record)) 
FROM visits
limit 10; 

```
